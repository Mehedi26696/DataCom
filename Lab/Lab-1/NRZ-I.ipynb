<center><h1>NRZ-I Encoding and Decoding Python Impelementation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = '10110010'
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

<b>Plot Signal

```python
def plot_signal(encoded_signal):
    time = range(len(encoded_signal)+1)
    signal = encoded_signal + [encoded_signal[-1]]
    plt.step(time, signal, where='post')
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-I Input')
    plt.grid(True)
    plt.savefig("NRZ-I_input_plot.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
plot_signal(bits_list)
```

<b>Encoding

```python
def nrzi_encode(bits,initial_state):
    encoded_signal=''
    previous_state = initial_state
    for bit in bits:
        if bit == '1':
            if previous_state == '0':
                previous_state = '1'
            else:
                previous_state = '0'
        # No change for 0, so just append the current level
        encoded_signal+=previous_state
    return encoded_signal
```

<b>Initial Low

```python
encoded_signal_low = nrzi_encode(bits,'0')
print(f"Encoded Signal: {encoded_signal_low}")
```

```python
encoded_signal_low__list = [int(char) for char in encoded_signal_low]
print(encoded_signal_low__list)
```

<b>Plot Signal

```python
def plot_signal_encode_low(encoded_signal):
    time = range(len(encoded_signal)+1)
    signal = encoded_signal + [encoded_signal[-1]]
    plt.step(time, signal, where='post')
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-I Encoding Initial Low')
    plt.grid(True)
    plt.savefig("NRZ-I_Encoding.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
plot_signal_encode_low(encoded_signal_low__list)
```

<b>Decoding

```python
def nrzi_decode(encoded_signal):
    decoded_data = '1'  # The first bit is assumed to be 1 (since initial level is arbitrary)
   # decoded_data = '0'
    
    for i in range(1, len(encoded_signal)):
        if encoded_signal[i] == encoded_signal[i - 1]:
            decoded_data += '0'  # No transition means 0
        else:
            decoded_data += '1'  # Transition means 1

    return decoded_data
```

<b>Initial Low

```python
print(f"Encoded Signal: {encoded_signal_low}")
decoded_signal_low = nrzi_decode(encoded_signal_low)
print(f"Decoded Signal: {decoded_signal_low}")
```

```python
decoded_signal_low_list = [int(char) for char in decoded_signal_low]
print(decoded_signal_low_list)
```
