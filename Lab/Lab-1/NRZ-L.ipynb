<center><h1>NRZ-L Encoding and Decoding Python Impelementation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = '10111001'
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

<b>Plot Signal

```python
def plot_signal(encoded_signal):
    time = range(len(encoded_signal))
    plt.step(time, encoded_signal, where='post', label="NRZ-I Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-L Input')
    plt.grid(True)
    plt.show()
```

```python
plot_signal(bits_list)
```

<b>Encoding

```python
def nrz_l_encode(bits):
    # NRZ-L encoding: 1 -> +1, 0 -> 0
    encoded_signal =[]
    for bit in bits:
        if bit == '1':
            encoded_signal.append(1)
        else:
            encoded_signal.append(0)
    return encoded_signal
```

```python
encoded_signal = nrz_l_encode(bits)
print(f"Encoded Signal: {encoded_signal}")
```

<b>Plot Signal

```python
def plot_signal_encoding(encoded_data):
    # Time vector for plotting
    time = np.linspace(0, len(encoded_data), len(encoded_data) * 100)
    signal = np.repeat(encoded_data, 100)

    # Plot the signal
    plt.step(time, signal, where='post', label="NRZ-L")
    plt.title("NRZ-L Encoding")
    plt.xlabel("Time")
    plt.ylabel("Voltage")
    plt.grid(True)
    plt.show()
```

```python
plot_signal_encoding(encoded_signal)
```

<b>Decoding

```python
def nrz_l_decode(encoded_data):
    # NRZ-L decoding: +1 -> 1, 0 -> 0
    decoded_data = ''
    for value in encoded_data:
        if value == 1:
            decoded_data += '1'
        else:
            decoded_data += '0'
    return decoded_data
```

```python
decoded_data = nrz_l_decode(encoded_signal)
print(f"Decoded Signal: {decoded_data}")
```

```python
decoded_signal= [int(char) for char in decoded_data]
print(decoded_signal)
```

<b>Plot Signal

```python
def plot_signal_decode(encoded_signal):
    time = range(len(encoded_signal))
    plt.step(time, encoded_signal, where='post', label="NRZ-I Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-L Decoding high')
    plt.grid(True)
    plt.show()
```

```python
plot_signal_decode(decoded_signal)
```
