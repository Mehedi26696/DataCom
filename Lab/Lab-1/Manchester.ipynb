```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits= "10111001"
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

<b>Plot Signal

```python
def plot_signal(encoded_signal):
    time = range(len(encoded_signal))
    plt.step(time, encoded_signal, where='post', label="NRZ-I Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('Manchester Input')
    plt.grid(True)
    plt.show()
```

```python
plot_signal(bits_list)
```

<b>Standard Manchester</b>
- For 1 low to high (0->1)
- For 0 high to low(1->0)

<b>Encoding

```python
def manchester_encode(bits):
    encoded = []
    for bit in bits:
        if bit == '1':
            # 1 is represented by Low to High transition (0 → 1)
            encoded.append('01')  # '0' → '1'
        else:
            # 0 is represented by High to Low transition (1 → 0)
            encoded.append('10')  # '1' → '0'
    return ''.join(encoded)
```

```python
encoded_signal_standard = manchester_encode(bits)
print(f"Encoded Signal: {encoded_signal_standard}")
```

<b>Plot Signal

```python
encoded_signal_list = [int(char) for char in encoded_signal_standard]
print(encoded_signal_list)
```

```python
def plot_signal(signal, title="Manchester Encoded Signal"):
    time = np.arange(len(signal))/2
    plt.figure(figsize=(10, 3))
    plt.step(time, signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.yticks([0, 1])
    plt.grid(True)
    plt.show()
```

```python
plot_signal(encoded_signal_list)
```

<b>Alternate Manchester</b>
- For 1 high to low (1->0)
- For 0 low to high(0->1)

```python
def manchester_alternate_encode(bits):
    encoded = []
    for bit in bits:
        if bit == '1':
            # 1 is represented by high to low transition (1 -> 0)
            encoded.append('10') # '1' → '0'
        else:
            # 0 is represented by low to high transition (0 -> 1)
            encoded.append('01') # '0' → '1'  
    return ''.join(encoded)
```

```python
bits= "10111001"
```

```python
encoded_signal_alternate = manchester_alternate_encode(bits)
print(f"Encoded Signal: {encoded_signal_alternate}")
```

```python
encoded_signal_list = [int(char) for char in encoded_signal_alternate]
print(encoded_signal_list) 
```

```python
def plot_signal(signal, title="Manchester Encoded Signal"):
    time = np.arange(len(signal))/2
    plt.figure(figsize=(10, 3))
    plt.step(time, signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.yticks([0, 1])
    plt.grid(True)
    plt.show()
```

```python
plot_signal(encoded_signal_list)
```

<b>Decode

<b>Standard

```python
def manchester_decode(encoded_signal):
    decoded_data =''
    for i in range(0,len(encoded_signal),2):
        if encoded_signal[i] == '0' and encoded_signal[i+1] == '1':
            decoded_data+='1'
        elif encoded_signal[i] == '1' and encoded_signal[i+1] == '0':
            decoded_data+='0'
    return decoded_data
```
