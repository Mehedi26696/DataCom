<center><h1>Pseudoternary Encoding and Decoding.</center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = "10110010"
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

<b>Input

```python
def plot_signal_input(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Input for:  {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_input(bits_list)
```

<b>Encoding

```python
def pseudo_encoder(bits,initial_state):
    pseudo_signal = []
    
    for bit in bits:
        if bit == '0':
            initial_state*= -1
            pseudo_signal.append(initial_state)
        else:
            pseudo_signal.append(0)
    return pseudo_signal
```

<b>Initial State Negative

```python
pseudo_encoded_signal_positive = pseudo_encoder(bits,1)
```

```python
print(f"Pseudoternary Encoded Signal Initial Positive : {pseudo_encoded_signal_positive}")
```

```python
def plot_signal_initial_positive(pseudo_encoded_signal_positive):
    time = range(len(pseudo_encoded_signal_positive)+1)
    signal = pseudo_encoded_signal_positive + [pseudo_encoded_signal_positive[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([-1, 0, 1])
    plt.grid(True)
    plt.title(f"Pseudoternary Encoding for Initial Positive: {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_initial_positive(pseudo_encoded_signal_positive)
```

<b>Initial State Negative

```python
pseudo_encoded_signal_negative = pseudo_encoder(bits,-1)
```

```python
print(f"Pseudoternary Encoded Signal Initial Negative : {pseudo_encoded_signal_negative}")
```

```python
def plot_signal_initial_negative(pseudo_encoded_signal_negative):
    time = range(len(pseudo_encoded_signal_negative)+1)
    signal = pseudo_encoded_signal_negative + [pseudo_encoded_signal_negative[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([-1, 0, 1])
    plt.grid(True)
    plt.title(f"Pseudoternary Encoding for Initial Negative: {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_initial_negative(pseudo_encoded_signal_negative)
```

<b>Decoding

```python
def pseudo_decode(pseudo_encoded_signal):
    decoded_signal =''
    for i in range(0,len(pseudo_encoded_signal)):
        if pseudo_encoded_signal[i] == 0:
            decoded_signal +='1'
        else:
            decoded_signal +='0'
        

    return decoded_signal
```

<b>Initial Positive

```python
pseudo_decoded_signal_positive = pseudo_decode(pseudo_encoded_signal_positive)
```

```python
print(f"Pseudo Decoded Signal Initial Positive : {pseudo_decoded_signal_positive}")
```

```python
decode_list_positive = [int(char) for char in pseudo_decoded_signal_positive]
print(decode_list_positive)
```

```python
def plot_signal_output_positive(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Output for positive:  {pseudo_decoded_signal_positive}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_output_positive(decode_list_positive)
```
