<center><h1>Pseudoternary encoding and decoding.</center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = "10110010"
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

```python
def plot_signal_input(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Input for:  {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_input(bits_list)
```

```python
def pseudo_encoder(bits):
    pseudo_signal = []
    last = 1

    for bit in bits:
        if bit == '0':
            last*= -1
            pseudo_signal.append(last)
        else:
            pseudo_signal.append(0)
    return pseudo_signal
```

```python
pseudo_encoded_signal = pseudo_encoder(bits)
```

```python
print(f"Pseudoternary Encoded Signal : {pseudo_encoded_signal}")
```

```python
def plot_signal(pseudo_encoded_signal):
    time = range(len(pseudo_encoded_signal)+1)
    signal = pseudo_encoded_signal + [pseudo_encoded_signal[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([-1, 0, 1])
    plt.grid(True)
    plt.title(f"Pseudoternary Encoding for: {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal(pseudo_encoded_signal)
```

```python
def pseudo_decode(pseudo_encoded_signal):
    decoded_signal =''
    for i in range(0,len(pseudo_encoded_signal)):
        if pseudo_encoded_signal[i] == 0:
            decoded_signal +='1'
        else:
            decoded_signal +='0'
        

    return decoded_signal
```

```python
pseudo_decoded_signal = pseudo_decode(pseudo_encoded_signal)
```

```python
print(f"Pseudo Decoded Signal : {pseudo_decoded_signal}")
```

```python
decode_list = [int(char) for char in pseudo_decoded_signal]
print(decode_list)
```

```python
def plot_signal_output(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Output for:  {pseudo_decoded_signal}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_output(decode_list)
```
