<center><h1>AMI Encoding</center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = "10110010"
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

```python
def plot_signal_input(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Input for:  {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_input(bits_list)
```

```python
def ami_encoder(bits):
    ami_signal = []
    last = -1

    for bit in bits:
        if bit == '1':
            last*= -1
            ami_signal.append(last)
        else:
            ami_signal.append(0)
    return ami_signal
```

```python
ami_encoded_signal = ami_encoder(bits)
```

```python
print(f"Ami Encoded Signal : {ami_encoded_signal}")
```

```python
def plot_signal(ami_encoded_signal):
    time = range(len(ami_encoded_signal)+1)
    signal = ami_encoded_signal + [ami_encoded_signal[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([-1, 0, 1])
    plt.grid(True)
    plt.title(f"AMI Encoding for: {bits}")
    plt.xlabel("Bit Index")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal(ami_encoded_signal)
```

```python
def ami_decode(ami_encoded_signal):
    decoded_signal =''
    for i in range(0,len(ami_encoded_signal)):
        if ami_encoded_signal[i] == 0:
            decoded_signal +='0'
        else:
            decoded_signal +='1'
        

    return decoded_signal

        
```

```python
ami_decoded_signal = ami_decode(ami_encoded_signal)
```

```python
print(f"Ami Decoded Signal : {ami_decoded_signal}")
```

```python
decode_list = [int(char) for char in ami_decoded_signal]
print(decode_list)
```

```python
def plot_signal_output(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Output for:  {ami_decoded_signal}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_output(decode_list)
```
