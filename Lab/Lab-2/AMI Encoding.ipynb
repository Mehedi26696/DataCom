<center><h1>AMI Encoding</center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = "10110010"
```

```python
bits_list = [int(char) for char in bits]
print(bits_list)
```

<b>Input Signal

```python
def plot_signal_input(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Input for:  {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_input(bits_list)
```

<b>Encoding

```python
def ami_encoder(bits,intial_state):
    ami_signal = []
    for bit in bits:
        if bit == '1':
            intial_state*= -1
            ami_signal.append(intial_state)
        else:
            ami_signal.append(0)
    return ami_signal
```

<b>Initial State Negative

```python
ami_encoded_signal_negative = ami_encoder(bits,-1)
```

```python
print(f"Ami Encoded Signal Initial Negative : {ami_encoded_signal_negative}")
```

```python
def plot_signal_negative(ami_encoded_signal_negative):
    time = range(len(ami_encoded_signal_negative)+1)
    signal = ami_encoded_signal_negative + [ami_encoded_signal_negative[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([-1, 0, 1])
    plt.grid(True)
    plt.title(f"AMI Encoding for intial negative: {bits}")
    plt.xlabel("Bit Index")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_negative(ami_encoded_signal_negative)
```

<b>Initial State Positive

```python
ami_encoded_signal_positive = ami_encoder(bits,1)
```

```python
print(f"Ami Encoded Signal Initial Positive : {ami_encoded_signal_positive}")
```

```python
def plot_signal_positive(ami_encoded_signal_negative):
    time = range(len(ami_encoded_signal_negative)+1)
    signal = ami_encoded_signal_negative + [ami_encoded_signal_negative[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([-1, 0, 1])
    plt.grid(True)
    plt.title(f"AMI Encoding for intial positive: {bits}")
    plt.xlabel("Bit Index")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_positive(ami_encoded_signal_positive)
```

<b>Decoding

```python
def ami_decode(ami_encoded_signal):
    decoded_signal =''
    for i in range(0,len(ami_encoded_signal)):
        if ami_encoded_signal[i] == 0:
            decoded_signal +='0'
        else:
            decoded_signal +='1'
        

    return decoded_signal

        
```

<b>Decode Negative

```python
ami_decoded_signal_negative = ami_decode(ami_encoded_signal_positive)
```

```python
print(f"Ami Decoded Signal Negative : {ami_decoded_signal_negative}")
```

```python
decode_list_negative = [int(char) for char in ami_decoded_signal_negative]
print(decode_list_negative)
```

```python
def plot_signal_output_negative(bits_list):
    time = range(len(bits_list)+1)
    signal = bits_list + [bits_list[-1]]
    plt.step(time,signal,where="post")
    plt.xticks(np.arange(len(bits) + 1))
    plt.yticks([0, 1])
    plt.grid(True)
    plt.title(f"Output for negative:  {ami_encoded_signal_negative}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
plot_signal_output_negative(decode_list_negative)
```
